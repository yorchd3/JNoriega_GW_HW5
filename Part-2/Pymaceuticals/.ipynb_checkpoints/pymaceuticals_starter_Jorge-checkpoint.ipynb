```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)
trial_data.head()
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(trial_data, mouse_data, how="left", on="Mouse ID")
```

```python
# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
#reduced form
reduc_combined_data = combined_data[["Mouse ID", "Timepoint", "Tumor Volume (mm3)", "Drug"]]
reduc_combined_data.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumorvol = reduc_combined_data.groupby(["Drug","Timepoint"]).mean()
mean_tumorvol.head()
```

```python
# Convert to DataFrame
mean_df = pd.DataFrame(mean_tumorvol)
```

```python
# Preview DataFrame
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
std_errors = reduc_combined_data.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
std_errors = pd.DataFrame(std_errors)
std_errors = std_errors.rename(columns = {"Tumor Volume (mm3)":"Standard Errors"})
std_errors = std_errors

# Preview DataFrame
std_errors.head()

#pivot errors to add to chart below
std_errors_piv = std_errors.pivot_table(index = "Timepoint", columns="Drug", values = "Standard Errors")
std_errors_piv.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reform_df = mean_df.pivot_table(index = "Timepoint", columns="Drug", values = "Tumor Volume (mm3)")
```

```python
# Preview that Reformatting worked
reform_df
```

```python
# Generate the Plot (with Error Bars)
time = np.arange(0, 50, 5)

fig, ax = plt.subplots()

Capomulin = ax.errorbar(time, reform_df.loc[:,"Capomulin"], yerr=std_errors_piv.loc[:,"Capomulin"],color="red", marker="o", linestyle = '--', linewidth = 0.5)
Infubinol = ax.errorbar(time, reform_df.loc[:,"Infubinol"], yerr=std_errors_piv.loc[:,"Infubinol"],color="blue", marker="^",linestyle = '--', linewidth = 0.5)
Ketapril = ax.errorbar(time, reform_df.loc[:,"Ketapril"], yerr=std_errors_piv.loc[:,"Ketapril"],color="green", marker = "s",linestyle = '--', linewidth = 0.5)
Placebo = ax.errorbar(time, reform_df.loc[:,"Placebo"], yerr=std_errors_piv.loc[:,"Placebo"],color="black", marker = "d",linestyle = '--', linewidth = 0.5)

# Create a legend for our chart
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")
# Show the chart
plt.grid()
```

```python
# Save the Figure
fig.savefig("../Images/Pharma_Tumor_ Jorge.png")
```

```python
# Show the Figure
fig.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
#reduced form
reduc_combined_data2 = combined_data[["Mouse ID", "Timepoint", "Metastatic Sites", "Drug"]]
#reduc_combined_data2.head()
#new data set
mean_metas = reduc_combined_data2.groupby(["Drug","Timepoint"]).mean()

# Convert to DataFrame
mean_metas = pd.DataFrame(mean_metas)

# Preview DataFrame
mean_metas.head()
```

```python
reform2_df = mean_metas.pivot_table(index = "Timepoint", columns="Drug", values = "Metastatic Sites")
reform2_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
std_errors2 = reduc_combined_data2.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
std_errors2 = pd.DataFrame(std_errors2)
std_errors2 = std_errors2.rename(columns = {"Metastatic Sites":"Standard Errors"})
std_errors2 = std_errors2

# Preview DataFrame
std_errors2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
std_errors2_piv = std_errors2.pivot_table(index = "Timepoint", columns="Drug", values = "Standard Errors")
std_errors2_piv = pd.DataFrame(std_errors2_piv)
# Preview that Reformatting worked
std_errors2_piv.head()
```

```python
# Generate the Plot (with Error Bars)
time = np.arange(0, 50, 5)

fig2, ax = plt.subplots()

Capomulin = ax.errorbar(time, reform2_df.loc[:,"Capomulin"], yerr=std_errors2_piv.loc[:,"Capomulin"],color="red", marker="o", linestyle = '--', linewidth = 0.5)
Infubinol = ax.errorbar(time, reform2_df.loc[:,"Infubinol"], yerr=std_errors2_piv.loc[:,"Infubinol"],color="blue", marker="^",linestyle = '--', linewidth = 0.5)
Ketapril = ax.errorbar(time, reform2_df.loc[:,"Ketapril"], yerr=std_errors2_piv.loc[:,"Ketapril"],color="green", marker = "s",linestyle = '--', linewidth = 0.5)
Placebo = ax.errorbar(time, reform2_df.loc[:,"Placebo"], yerr=std_errors2_piv.loc[:,"Placebo"],color="black", marker = "d",linestyle = '--', linewidth = 0.5)

# Create a legend for our chart
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.title("Metatastic Spread During Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Met. Sites")
# Show the chart
plt.grid()

# Save the Figure
fig2.savefig("../Images/Pharma_Metastatic_Jorge.png")

# Show the Figure
plt.plot()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
#reduced form
reduc_combined_data3 = combined_data[["Mouse ID", "Timepoint", "Drug"]]
reduc_combined_data3.head()
#new data set
mouse_count = reduc_combined_data3.groupby(["Drug","Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
mouse_count = pd.DataFrame(mouse_count)

# Preview DataFrame
mouse_count = mouse_count.rename(columns = {"Mouse ID": "Mouse Count"})
mouse_count.head()
```

```python
#mouse_count["Change"] = mouse_count.pct_change()
#mouse_count['Index'] = 100*np.exp(np.nan_to_num(mouse_count["Change"].cumsum()))
#mouse_count.head()
#mouse_count = mouse_count[["Index"]]
#mouse_count
```

```python
# Minor Data Munging to Re-Format the Data Frames
reform3_df = mouse_count.pivot_table(index = "Timepoint", columns="Drug", values = "Mouse Count")
reform3_df.head()
```

```python
reform3_df = reform3_df.pct_change()
reform3_df = 100*np.exp(np.nan_to_num(reform3_df.cumsum()))
reform3_df = pd.DataFrame(reform3_df)
#reform3_df


#columns = ["Capomulin","Ceftamin","Infubinol","Ketapril","Naftisol","Placebo","Propriva","Ramicane","Stelasyn","Zoniferol"]

reform3_df = reform3_df.rename(columns = {0: "Capomulin", 2:"Infubinol", 3:"Katapril", 5:"Placebo"})
reform3_df
```

```python
# Generate the Plot (Accounting for percentages)
time = np.arange(0, 50, 5)

Capomulin, = plt.plot(time, reform3_df.loc[:,"Capomulin"],color="red", marker="o", linestyle = '--', linewidth = 0.5)
Infubinol, = plt.plot(time, reform3_df.loc[:,"Infubinol"],color="blue", marker="^",linestyle = '--', linewidth = 0.5)
Ketapril, = plt.plot(time, reform3_df.loc[:,"Katapril"], color="green", marker = "s",linestyle = '--', linewidth = 0.5)
Placebo, = plt.plot(time, reform3_df.loc[:,"Placebo"], color="black", marker = "d",linestyle = '--', linewidth = 0.5)

# Create a legend for our chart
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Survival Rate %")
# Show the chart
plt.grid()

# Save the Figure
plt.savefig("../Images/Pharma_Survival_Jorge.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
initial = reform_df.iloc[[0,9]]
initial = initial.pct_change()*100
initial = initial.dropna()

# Display the data to confirm
initial
```

```python
# Store all Relevant Percent Changes into a Tuple
per_chg = ()
per_chg = initial
per_chg
```

```python
N = 5
men_means = (20, 35, 30, 35, 27)
men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)

women_means = (25, 32, 34, 20, 25)
women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='y', yerr=women_std)

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))
print(ind + width/2)


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
# Orient widths. Add labels, tick marks, etc. 
N = 4
ind = np.arange(N) 
width = 1

fig, ax = plt.subplots()
Capomulin = ax.bar(ind[0], per_chg["Capomulin"], width, color='green', align="edge")
Infubinol = ax.bar(ind[1], per_chg["Infubinol"], width, color='red', align="edge")
Ketapril = ax.bar(ind[2], per_chg["Ketapril"], width, color='red', align="edge")
Placebo = ax.bar(ind[3], per_chg["Placebo"], width, color='red', align="edge")

ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change over 45 day Treatment')

# Create a list indicating where to write x labels and set figure size to adjust for space
ax.set_xticks(ind + 1)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
plt.ylim(-30, 70)
plt.grid()

# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 0.2*height,'%d' % int(height)+"%", color = "white", ha='center', va='center')

# Call functions to implement the function calls
autolabel(Capomulin)
autolabel(Infubinol)
autolabel(Ketapril)
autolabel(Placebo)

# Save the Figure


# Show the Figure
fig.show()
```
